## Cumulative Distribution Functions

### Percentiles

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import nsfg
import pandas as pd
import random
import thinkstats2
import thinkplot
```

```python
def percentileRank(scores,your_score):
    count=0
    for score in scores:
        if score <= your_score:
            count+=1
            
    percentile_rank=100.0* count/len(scores)
    return percentile_rank

def percentile(scores,percentile_rank):
    scores.sort()
    index=int(percentile_rank * (len(scores)-1)/100)
    return scores[index]


scores = [ random.randint(0, 30) for i in range(30) ]
score=10

rank=percentileRank(scores,score)
percentile=percentile(scores,rank)
print(rank)
print(percentile)
    
    
```

```python
def evalCDF(scores,your_score):
    count=0
    for score in scores:
        if score <= your_score:
            count+=1
            
    probability=count/len(scores)
    return probability

probabilities=[]

for score in scores:
    prob=evalCDF(scores,score)
    probabilities.append(prob)
    
print(scores)
    
plt.step(
    scores,
    probabilities
)
plt.title("CDF Step Function Graph")
plt.xlabel("Score")
plt.ylabel("Probability")
    
```

```python
df=nsfg.ReadFemPreg()

firsts=df[df['pregordr']==1  & (df["prglngth"]>27) & (df["prglngth"]<46)]
others=df[df['pregordr']!=1  & (df["prglngth"]>27) & (df["prglngth"]<46)]


firstTWeights=list(firsts["totalwgt_lb"].dropna())
othersTWeights=list(others["totalwgt_lb"].dropna())

firstTWeights.sort()
othersTWeights.sort()




firtsProbabilities=[]
for weight in firstTWeights:
    prob=evalCDF(firstTWeights,weight)
    firtsProbabilities.append(prob)
    
othersProbabilities=[]
for weight in othersTWeights:
    prob=evalCDF(othersTWeights,weight)
    othersProbabilities.append(prob)
    

plt.figure(figsize=(15,10))    
    
    


plt.step(
    firstTWeights,
    firtsProbabilities,
    label="First Pregnancies Weights"
)

plt.step(
    othersTWeights,
    othersProbabilities,
    label="Other Pregnancies total Weights"
    
)
plt.title("CDF Step Function Graph")
plt.xlabel("Score")
plt.ylabel("Probability")
plt.legend()






```
